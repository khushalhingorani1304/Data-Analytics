```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,52]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.title('weather')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label="Avg")
plt.legend(loc ="best",shadow=True,fontsize="large")
```

lecture 4

```python
company = ['Google','Amazon','Microsoft','FaceBook']
revenue = [90,136,89,27]
profit = [40,2,96,54]
```

```python
ypos = np.arange(len(company))
```

```python
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,label='Revenue',width=0.4)
plt.bar(ypos+0.2,profit,label='Profit',width=0.4)
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.barh(ypos-0.2,revenue,label='Revenue')
plt.barh(ypos+0.2,profit,label='Profit')
plt.legend()
```

HISTOGRAM

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [103,89,80,160,139,84,119,115,125,90,67,87,109]
# plt.xlabel('Sugar analysis')
# plt.ylabel('No. of Patients')
# plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth = 0.95,color=['green','orange'],label=['men','women'])
plt.legend()
```
